# 도서관정보나루 LIBRARY BIGDATA OPEN API

### 인기대출 도서 조회

```
import json
```

```
url = "http://data4library.kr/api/loanItemSrch?authKey=[인증키]&startDt=2026-01-01&endDt=2026-04-08&gender=0&pageNo=1&pageSize=100&format=json"
```

```
res = requests.get(url)
text = res.text
BD_json = json.loads(text) # BD = BOOK DATA
```

```
print(json.dumps(BD_json, indent=4, sort_keys=True))
```
